```python
import seaborn as sns
import matplotlib.pyplot as plt

# Función para graficar la matriz de correlación con parámetros reutilizables
def graficar_matriz_correlacion(correlaciones, titulo):
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlaciones, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
    plt.title(titulo)
    plt.show()

# Gráfico de la matriz de correlación 2004
graficar_matriz_correlacion(correlaciones_2004_alt, 'Matriz de correlación 2004')

# Gráfico de la matriz de correlación 2024
graficar_matriz_correlacion(correlaciones_2024_alt, 'Matriz de correlación 2024')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Función para agregar la columna PET
def calcular_pet(edad):
    return 1 if 15 <= edad <= 65 else 0

# Agregamos la columna PET
respondieron['PET'] = respondieron['ch06'].apply(calcular_pet)

# Filtramos los datos por años 2004 y 2024
def filtrar_datos_por_ano(dataframe, ano):
    return dataframe[dataframe['ano4'] == ano]

respondieron_2004 = filtrar_datos_por_ano(respondieron, 2004)
respondieron_2024 = filtrar_datos_por_ano(respondieron, 2024)

# Contamos la cantidad de personas en PET para 2004 y 2024
def contar_pet(dataframe):
    return dataframe['PET'].value_counts()

pet_2004 = contar_pet(respondieron_2004)
pet_2024 = contar_pet(respondieron_2024)

# Crear el gráfico de barras comparando los dos años
def crear_grafico_barras(pet_2004, pet_2024):
    labels = ['SET 2004', 'PET 2004', 'SET 2024', 'PET 2024']
    valores = [pet_2004.get(0, 0), pet_2004.get(1, 0), pet_2024.get(0, 0), pet_2024.get(1, 0)]
    
    fig, ax = plt.subplots()
    ax.bar(labels, valores, color=['red', 'yellow', 'red', 'yellow'])

    ax.set_ylabel('Número de personas')
    ax.set_title('Composición de la Población en Edad para Trabajar (PET) y SET en 2004 y 2024')
    plt.show()

# Crear el gráfico
crear_grafico_barras(pet_2004, pet_2024)

# Comentarios
# - En ambos años hay menos personas sin edad para trabajar (rojo).
# - Las diferencias en SET y PET son similares en los dos años, es decir, parece haber una proporción similar.
# - Hay menos datos en 2024.
```

```python
#EJERCICIO 2 - 2004

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import confusion_matrix, roc_curve, auc, accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Lista para almacenar resultados de cada modelo
results = []

# Función para entrenar, predecir y obtener métricas de un modelo
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Matriz de confusión
    conf_matrix = confusion_matrix(y_test, y_pred)

    # Curva ROC y AUC
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    # Accuracy y AUC
    accuracy = accuracy_score(y_test, y_pred)
    auc_score = roc_auc_score(y_test, y_prob)

    # Guardamos resultados
    results.append({"Model": model_name, "Confusion Matrix": conf_matrix, "Accuracy": accuracy, "AUC": auc_score})

    # Graficamos la curva ROC
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')

# Inicializamos gráficos
plt.figure(figsize=(10, 8))

# Imputador para manejar NaN
imputer = SimpleImputer(strategy='mean')

# Pipeline para Regresión Logística
pipeline_logistic = Pipeline([
    ('imputer', imputer),
    ('classifier', LogisticRegression())
])
evaluate_model(pipeline_logistic, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "Regresión Logística 2004")

# Pipeline para Análisis Discriminante Lineal
pipeline_lda = Pipeline([
    ('imputer', imputer),
    ('classifier', LinearDiscriminantAnalysis())
])
evaluate_model(pipeline_lda, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "Análisis Discriminante Lineal 2004")

# Pipeline para KNN
pipeline_knn = Pipeline([
    ('imputer', imputer),
    ('classifier', KNeighborsClassifier(n_neighbors=3))
])
evaluate_model(pipeline_knn, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "KNN (k=3) 2004")

# Pipeline para Naive Bayes
pipeline_naive_bayes = Pipeline([
    ('imputer', imputer),
    ('classifier', GaussianNB())
])
evaluate_model(pipeline_naive_bayes, X_train_2004, y_train_2004, X_test_2004, y_test_2004, "Naive Bayes 2004")

# Configuración de la gráfica de ROC
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC')
plt.legend(loc="lower right")
plt.show()

# Imprimir resultados
for result in results:
    print(f"Modelo: {result['Model']}")
    print("Matriz de Confusión:\n", result['Confusion Matrix'])
    print("Accuracy:", result['Accuracy'])
    print("AUC:", result['AUC'])
    print("\n" + "-"*30 + "\n")
```
